
# Analytic \(P_s(\alpha,d,\kappa)\) fit, train/verification split, and loss heat maps

This notebook fits the three free empirical parameters \(a,b,q\) in

\[
P_s=
\frac{1}{4\pi^2\alpha}
\exp\!\left[
-\frac{2(1+d)}{d}
+\frac{2}{d}I(d,\kappa)
+a+b\,d
\right]
\left[C(\nu_\star)\right]^q ,
\]

then checks the fitted formula on a held-out verification set.

The split is done by unique \((d,\kappa)\) pairs, so all sampled \(\alpha\)-values for a given pair stay either in training or verification.


```python

from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy import sparse
from scipy.optimize import linprog
from scipy.special import gammaln

CSV_PATH = Path("genesis_scan_table(12).csv")
if not CSV_PATH.exists():
    CSV_PATH = Path("/mnt/data/genesis_scan_table(12).csv")

TRAIN_FRAC = 0.80
SEED = 1234
TAU_B = 1.0
```

```python

def log_base_and_logc(alpha, d, kappa, tau_b=1.0):
    # Returns log(P_base) and log(C(nu)) for
    # log(Ps) = log(P_base) + a + b*d + q*log(C(nu)).
    # Uses the closed-form integral I(d,kappa).
    A = 3.0 * (1.0 + d) ** 2
    s = np.exp(d * np.log(kappa))                     # kappa**d
    nu = np.sqrt(2.25 + A * s * (1.0 - s))

    u = A * (1.0 - s) / (nu + 1.5)                    # stable form
    rootA = np.sqrt(A)

    I = (
        3.0
        + rootA * np.arctan(u / rootA)
        + 1.5 * np.log(A / (A + u**2))
        - A * (3.0 + u) / (A + u**2)
    )

    log_base = (
        -np.log(4.0 * np.pi**2 * tau_b**2 * alpha)
        - 2.0 * (1.0 + d) / d
        + 2.0 * I / d
    )

    logC = (2.0 * nu - 3.0) * np.log(2.0) + 2.0 * (
        gammaln(nu) - gammaln(1.5)
    )
    return log_base, logC


def predict_log_ps(df, params, tau_b=1.0):
    # Formula prediction in log-space.
    a, b, q = params
    alpha = df["alpha"].to_numpy(float)
    d = df["d"].to_numpy(float)
    kappa = df["kappa"].to_numpy(float)

    log_base, logC = log_base_and_logc(alpha, d, kappa, tau_b=tau_b)
    return log_base + a + b * d + q * logC
```

```python

def split_by_d_kappa(df, train_frac=0.8, seed=1234):
    # Random split by unique (d,kappa), keeping all alpha rows together.
    pair_id = pd.factorize(pd.MultiIndex.from_frame(df[["d", "kappa"]]))[0]
    unique_ids = np.arange(pair_id.max() + 1)

    rng = np.random.default_rng(seed)
    rng.shuffle(unique_ids)

    train_ids = unique_ids[: int(train_frac * len(unique_ids))]
    train_mask = np.isin(pair_id, train_ids)
    return df.loc[train_mask].copy(), df.loc[~train_mask].copy()


def fit_minimax_log_error(df, tau_b=1.0):
    # Fits a,b,q by minimizing max absolute log-error:
    #     minimize t subject to |X @ [a,b,q] - y| <= t.
    alpha = df["alpha"].to_numpy(float)
    d = df["d"].to_numpy(float)
    kappa = df["kappa"].to_numpy(float)
    log_ps = np.log(df["Ps"].to_numpy(float))

    log_base, logC = log_base_and_logc(alpha, d, kappa, tau_b=tau_b)
    y = log_ps - log_base

    n = len(df)
    X = sparse.csr_matrix(np.column_stack([np.ones(n), d, logC]))
    minus_t = sparse.csr_matrix(-np.ones((n, 1)))

    A_ub = sparse.vstack(
        [sparse.hstack([X, minus_t]), sparse.hstack([-X, minus_t])],
        format="csr",
    )
    b_ub = np.concatenate([y, -y])

    res = linprog(
        c=np.array([0.0, 0.0, 0.0, 1.0]),
        A_ub=A_ub,
        b_ub=b_ub,
        bounds=[(None, None), (None, None), (None, None), (0.0, None)],
        method="highs",
    )
    if not res.success:
        raise RuntimeError(res.message)

    return res.x[:3], res.x[3]


def evaluate(df, params, tau_b=1.0):
    # Returns max-error diagnostics and summary loss statistics.
    log_true = np.log(df["Ps"].to_numpy(float))
    log_pred = predict_log_ps(df, params, tau_b=tau_b)

    rel_error = np.expm1(log_pred - log_true)
    abs_rel_error = np.abs(rel_error)
    i = int(np.argmax(abs_rel_error))
    row = df.iloc[i]

    return {
        "rows": len(df),
        "max_abs_relative_error": abs_rel_error[i],
        "max_abs_relative_error_percent": 100.0 * abs_rel_error[i],
        "ratio_formula_over_numeric_at_max": np.exp(log_pred[i] - log_true[i]),
        "alpha_at_max": row["alpha"],
        "d_at_max": row["d"],
        "kappa_at_max": row["kappa"],
        "Ps_numeric_at_max": row["Ps"],
        "Ps_formula_at_max": np.exp(log_pred[i]),
        "mean_abs_error_percent": 100.0 * abs_rel_error.mean(),
        "median_abs_error_percent": 100.0 * np.median(abs_rel_error),
    }
```

```python

df = pd.read_csv(CSV_PATH, usecols=["alpha", "d", "kappa", "Ps"])

train, verify = split_by_d_kappa(df, train_frac=TRAIN_FRAC, seed=SEED)
params, train_minimax_log_error = fit_minimax_log_error(train, tau_b=TAU_B)

a, b, q = params
train_stats = evaluate(train, params, tau_b=TAU_B)
verify_stats = evaluate(verify, params, tau_b=TAU_B)

print(f"Loaded rows: {len(df):,}")
print(f"Training rows: {len(train):,}")
print(f"Verification rows: {len(verify):,}")
print()
print("Fitted parameters:")
print(f"a = {a:.12g}")
print(f"b = {b:.12g}")
print(f"q = {q:.12g}")
print()
print("Fitted formula:")
print("Ps = P_base * exp(a + b*d) * C(nu)^q")
```

```python

stats = pd.DataFrame(
    {
        "training": train_stats,
        "verification": verify_stats,
    }
)
stats
```

```python

def pair_loss_table(df_eval, params, tau_b=1.0):
    # Max |relative error| in percent over alpha for each (d,kappa) pair.
    tmp = df_eval[["alpha", "d", "kappa", "Ps"]].copy()
    log_true = np.log(tmp["Ps"].to_numpy(float))
    log_pred = predict_log_ps(tmp, params, tau_b=tau_b)

    tmp["loss_percent"] = 100.0 * np.abs(np.expm1(log_pred - log_true))
    return (
        tmp.groupby(["d", "kappa"], as_index=False)["loss_percent"]
        .max()
        .assign(log10_kappa=lambda x: np.log10(x["kappa"]))
    )


def plot_verification_loss_heatmap(pair_loss, output_path=None):
    # Random pair splitting makes the held-out grid sparse, so points are preferable.
    fig, ax = plt.subplots(figsize=(9.5, 5.4), dpi=140)

    sc = ax.scatter(
        pair_loss["log10_kappa"],
        pair_loss["d"],
        c=pair_loss["loss_percent"],
        s=24,
    )
    cbar = fig.colorbar(sc, ax=ax)
    cbar.set_label("max |relative error| over alpha (%)")

    ax.set_xlabel(r"$\log_{10}\kappa$")
    ax.set_ylabel(r"$d$")
    ax.set_title("Verification loss heat map")
    ax.grid(True, linewidth=0.3, alpha=0.4)

    fig.tight_layout()
    if output_path is not None:
        fig.savefig(output_path, bbox_inches="tight")
    return fig, ax


verify_pair_loss = pair_loss_table(verify, params, tau_b=TAU_B)
fig, ax = plot_verification_loss_heatmap(
    verify_pair_loss,
    output_path="/mnt/data/verification_loss_heatmap.png",
)
plt.show()
```

```python

def plot_full_grid_loss_heatmap(df_eval, params, output_path=None, tau_b=1.0):
    # Dense heat map over the full (d,kappa) grid, max over alpha.
    pair_loss = pair_loss_table(df_eval, params, tau_b=tau_b)
    grid = pair_loss.pivot(index="d", columns="log10_kappa", values="loss_percent")

    fig, ax = plt.subplots(figsize=(9.5, 5.4), dpi=140)
    im = ax.imshow(
        grid.to_numpy(),
        origin="lower",
        aspect="auto",
        extent=[
            grid.columns.min(),
            grid.columns.max(),
            grid.index.min(),
            grid.index.max(),
        ],
    )
    cbar = fig.colorbar(im, ax=ax)
    cbar.set_label("max |relative error| over alpha (%)")

    ax.set_xlabel(r"$\log_{10}\kappa$")
    ax.set_ylabel(r"$d$")
    ax.set_title("Full-grid loss heat map using fitted parameters")

    fig.tight_layout()
    if output_path is not None:
        fig.savefig(output_path, bbox_inches="tight")
    return fig, ax


fig, ax = plot_full_grid_loss_heatmap(
    df,
    params,
    output_path="/mnt/data/full_grid_loss_heatmap.png",
    tau_b=TAU_B,
)
plt.show()
```
